1..write a python program to import iris dataset from sklearn and convert it into data frame and display the shape of the data set?

```python
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt
digits = load_digits()
print("Dataset shape:", digits.data.shape)
print("Image shape:", digits.images.shape)
print("Target names:", digits.target_names)
plt.figure(figsize=(4, 4))
plt.imshow(digits.images[5], cmap='gray')
plt.title(f"Label: {digits.target[5]}")
plt.axis("off")
plt.show()
```

```python
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt

digits = load_digits()

fig, axes = plt.subplots(5, 2, figsize=(10, 5))

for i, ax in enumerate(axes.flat):
    ax.imshow(digits.images[i], cmap='gray')
    ax.set_title(f"Digit: {digits.target[i]}")
    ax.axis('off')

plt.tight_layout()
plt.show()
```

```python

import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()

df = pd.DataFrame(iris.data, columns=iris.feature_names)
df["target"] = iris.target

print("First 5 rows of the DataFrame:")
print(df.head())
print("-" * 30)
print(f"Shape of the dataset: {df.shape}")
```

Write a python program to count number of samples available in iris dataset and digit dataset?


```python
from sklearn.datasets import load_digits, load_iris

iris = load_iris()
digits = load_digits()

iris_samples = iris.data.shape[0]
digits_samples = digits.data.shape[0]

print(f"Number of samples available in the Iris dataset: {iris_samples}")
print(f"Number of samples available in the Digits dataset: {digits_samples}")
```
